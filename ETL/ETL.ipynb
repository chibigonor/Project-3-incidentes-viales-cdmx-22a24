```python
import pandas as pd
```

```python
# First read the original CSV file in order to get the original dataframe
original_df = pd.read_csv('../Original_CSV/inViales_2022_2024.csv')
original_df.head()
```

```python
#In this section we create a new csv for alcaldias
alcaldias = original_df[['alcaldia_catalogo']].drop_duplicates().rename(columns={'alcaldia_catalogo': 'alcaldia'}).reset_index(drop=True).dropna()
alcaldias['id'] = alcaldias.index + 1
alcaldias = alcaldias[['id', 'alcaldia']]


alcaldias.to_csv('../Mod_CSV/alcaldias.csv', index=False)
alcaldias
```

```python
# we drop some alcaldias columns that weren't helpful
original_df_1 = original_df.drop(columns=['alcaldia_inicio','alcaldia_cierre', 'tipo_incidente_c4'])
original_df_1 = original_df_1[original_df_1['incidente_c4'] != 'Persona atropellada']
original_df_1
```

```python
#In this section we create a new csv for alcaldias
tipoincidente = original_df_1[['incidente_c4']].drop_duplicates().rename(columns={'incidente_c4': 'tipo_incidente'}).reset_index(drop=True).dropna()
tipoincidente['id'] = tipoincidente.index + 1
tipoincidente = tipoincidente[['id', 'tipo_incidente']]


tipoincidente.to_csv('../Mod_CSV/tipoincidente.csv', index=False)
tipoincidente
```

```python
# We use the count in order to clean the data
tipo_incidente_counts = original_df_1['incidente_c4'].value_counts()
tipo_incidente_counts_df = tipo_incidente_counts.reset_index().rename(columns={'index': 'tipo_incidente', 'incidente_c4': 'cantidad'})
tipo_incidente_counts_df
```

```python
colonias = original_df[['colonia_catalogo']].drop_duplicates().rename(columns={'colonia_catalogo': 'colonia'}).reset_index(drop=True).dropna()
colonias['id'] = colonias.index + 1
colonias = colonias[['id', 'colonia']]
colonias.to_csv('../Mod_CSV/colonias.csv', index=False)
colonias
```

```python
codigo_cierre = original_df[['codigo_cierre']].drop_duplicates().rename(columns={'codigo_cierre': 'codigo'}).reset_index(drop=True).dropna() 
codigo_cierre['id'] = codigo_cierre.index + 1
descripcion_mapping = { 'A': 'Afirmativo', 'N': 'Negativo', 'I': 'Informativo', 'D': 'Duplicados', 'F': 'Falso' }
codigo_cierre['descripcion'] = codigo_cierre['codigo'].map(descripcion_mapping)
codigo_cierre = codigo_cierre[['id', 'codigo', 'descripcion']]
codigo_cierre.to_csv('../Mod_CSV/codigo_cierre.csv', index=False)
codigo_cierre
```

```python
tipo_entrada = original_df[['tipo_entrada']].drop_duplicates().rename(columns={'tipo_entrada': 'tipo_entrada'}).reset_index(drop=True).dropna() 
tipo_entrada['id'] = tipo_entrada.index + 1
tipo_entrada = tipo_entrada[['id', 'tipo_entrada']]
tipo_entrada.to_csv('../Mod_CSV/tipo_entrada.csv', index=False)
tipo_entrada
```

```python
clasificacion_alarma = original_df[['clas_con_f_alarma']].drop_duplicates().rename(columns={'clas_con_f_alarma': 'clasificacion'}).reset_index(drop=True).dropna()
clasificacion_alarma['id'] = clasificacion_alarma.index + 1
clasificacion_alarma = clasificacion_alarma[['id', 'clasificacion']]
clasificacion_alarma.to_csv('../Mod_CSV/clasificacion_alarma.csv', index=False)
clasificacion_alarma
```

```python
incidentes = original_df_1.copy()
incidentes = incidentes.merge(alcaldias[['id', 'alcaldia']], how='left', left_on='alcaldia_catalogo', right_on='alcaldia').rename(columns={'id': 'alcaldia_catalogo_id'})
incidentes = incidentes.merge(tipoincidente[['id', 'tipo_incidente']], how='left', left_on='incidente_c4', right_on='tipo_incidente').rename(columns={'id': 'tipo_incidente_id'})
incidentes = incidentes.merge(codigo_cierre[['id', 'codigo']], how='left', left_on='codigo_cierre', right_on='codigo').rename(columns={'id': 'codigo_cierre_id'})
incidentes = incidentes.merge(clasificacion_alarma[['id', 'clasificacion']], how='left', left_on='clas_con_f_alarma', right_on='clasificacion').rename(columns={'id': 'clasificacion_alarma_id'})
incidentes = incidentes.merge(tipo_entrada[['id', 'tipo_entrada']], how='left', left_on='tipo_entrada', right_on='tipo_entrada').rename(columns={'id': 'tipo_entrada_id'})
incidentes = incidentes.merge(colonias[['id', 'colonia']], how='left', left_on='colonia_catalogo', right_on='colonia').rename(columns={'id': 'colonia_catalogo_id'})

incidentes = incidentes[['folio', 'fecha_creacion', 'hora_creacion', 'dia_semana', 'fecha_cierre', 'hora_cierre', 'tipo_incidente_id', 'codigo_cierre_id', 'clasificacion_alarma_id', 'tipo_entrada_id', 'alcaldia_catalogo_id', 'colonia_catalogo_id', 'longitud', 'latitud']]

#incidentes.to_csv('../Mod_CSV/incidentes.csv', index=False)

incidentes = incidentes[incidentes['colonia_catalogo_id'].isin(colonias['id'])]

#Code for filling na values
incidentes['tipo_entrada_id'] = incidentes['tipo_entrada_id'].fillna(1).astype(int)
incidentes['alcaldia_catalogo_id'] = incidentes['alcaldia_catalogo_id'].fillna(1).astype(int)
incidentes['colonia_catalogo_id'] = incidentes['colonia_catalogo_id'].fillna(1).astype(int)
incidentes['codigo_cierre_id'] = incidentes['codigo_cierre_id'].fillna(1).astype(int)
incidentes['clasificacion_alarma_id'] = incidentes['clasificacion_alarma_id'].fillna(1).astype(int)
incidentes['tipo_incidente_id'] = incidentes['tipo_incidente_id'].fillna(1).astype(int)

#Creating the csv
incidentes.to_csv('../Mod_CSV/incidentes.csv', index=False)

print("Incidentes")
print(incidentes.head())
```
